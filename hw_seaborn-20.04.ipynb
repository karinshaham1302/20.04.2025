```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

```python
df = pd.read_csv("mpg.csv")
df
```

```python
df_mpg = df.copy()
display(df_mpg.info())
display(df_mpg.describe().T)
```

```python
# 1.

plt.figure(figsize=(15,6))

sns.scatterplot(data=df_mpg, x='horsepower', y='mpg', hue='origin', palette='Spectral')

plt.tight_layout()
plt.legend()

plt.show()
```

```python
# 2.

plt.figure(figsize=(8,6))

df_cylinders = df_mpg[df_mpg['cylinders'].isin([4, 8])]
sns.barplot(data=df_cylinders, x='cylinders', y='mpg')
plt.title('Fuel Consumption Comparison: 4 vs 8 Cylinders')
plt.xlabel('Number of Cylinders')
plt.ylabel('mpg')

plt.show()
```

```python
# 3.

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

plt.figure(figsize=(10,4))

sns.countplot(data=df_mpg, x='origin', palette='Spectral')

plt.title('Number of Cars by Country of Origin')
plt.xlabel('Country of Origin')
plt.ylabel('Number of Cars')

plt.show()
```

```python
# 4.

plt.figure(figsize=(10,4))

sns.displot(data=df_mpg, x='weight', bins=30,  kde=True)

plt.title('Car Weights')

plt.show()
```

```python
# 5.

plt.figure(figsize=(10,4))

sns.boxplot(data=df_mpg, x='cylinders', y='mpg')

plt.title('MPG by Number of Cylinders')
plt.xlabel('Number of Cylinders')
plt.ylabel('MPG')

plt.show()
```
